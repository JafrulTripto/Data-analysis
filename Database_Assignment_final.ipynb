```python
import numpy as np
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
```

```python
File = pd.read_csv('dataset2_odi.csv') 
```

```python
stat=pd.DataFrame(File)
stat.columns=['Batsman','Runs','Minute','Balls','fours','sixes','StrikeRate','Position','Dismissal','Innings']
stat.head(10)
```

```python
b1 = stat[(stat.Batsman==1)].reset_index()
b2 = stat[(stat.Batsman==2)].reset_index()
b3 = stat[(stat.Batsman==3)].reset_index()
b4 = stat[(stat.Batsman==4)].reset_index()
b5 = stat[(stat.Batsman==5)].reset_index()
b6 = stat[(stat.Batsman==6)].reset_index()
```

```python
average= stat[['Batsman','Runs']].groupby('Batsman').mean() 
print(average)
average_runs=average.sum(axis = 1)
plt.ylabel('Average Runs')
plt.rcParams["figure.figsize"] = [8,7]
plt.show(average_runs.plot.bar())
```

```python
std= stat[['Batsman','Runs']].groupby('Batsman').std() 
print(std)
std_dev=std.sum(axis = 1)
plt.ylabel('Standerd deviation')
plt.rcParams["figure.figsize"] = [8,7]
plt.show(std_dev.plot.bar())
```

```python

b1['StrikeRate'].plot(color='Red',linewidth=4.0)
b2['StrikeRate'].plot( color='Black',linewidth=5.0)
b3['StrikeRate'].plot( color='Blue',linewidth=7.0)
b4['StrikeRate'].plot( color='Green',linewidth=6.0)
b5['StrikeRate'].plot( color='#b0b1b2',linewidth=9.0)
b6['StrikeRate'].plot( color='#00f4d0',linewidth=6.0)
plt.xlabel('Matches',fontsize=15)
plt.ylabel('Strike Rate',fontsize=15)
plt.rcParams["figure.figsize"] = [8,7]
plt.title('Matches VS Strike rate')
plt.show()
plt.gcf().clear()
plt.clf()
plt.cla()
plt.close()
```

```python
strike_rate= stat[['Batsman','StrikeRate']].groupby('Batsman').mean() 
print(strike_rate)               
plt.rcParams["figure.figsize"] = [8,7]
strike_rate.plot(kind='bar',color='green' ,title = " Average Strike rates of Batsman")
```

```python
sns.set()
_ = sns.swarmplot(x= 'Dismissal', y = 'Batsman', data = stat )
_ = plt.margins(0.1)
_ = plt.rcParams["figure.figsize"] = [8,7]
_ = plt.xlabel ('Dismissal')
_ = plt.ylabel ('Batsman')
plt.show()
```

```python
Boundaries= stat[['Batsman','fours','sixes']].groupby('Batsman').sum() #run from boundary from each batsman
plt.rcParams["figure.figsize"] = [10,7]
Boundaries.plot(kind='bar', title="Boundaries")
```

```python
Runs_VS_ball= stat[['Batsman','Balls','Runs']].groupby('Batsman').sum() 
Runs_VS_ball.plot(kind='bar' , title='Runs VS Balls')
```

```python
batsman1_Century= b1[(b1.Runs>=100)]
a=batsman1_Century.Runs.count()
print(('Batsman 1 centuries: '),a)
batsman1_HCentury= b1[(b1.Runs>=50) & (b1.Runs<100)]
b=batsman1_HCentury.Runs.count()
print(('Batsman 1 Half-centuries: '),b)
batsman1_below50= b1[(b1.Runs<50)]
c=batsman1_below50.Runs.count()
print(('Batsman 1 below Half-centuries: '),c)

labels = 'Century', 'Fifty', 'Below Fifty'
sizes = [a,b,c]
colors = ['gold', 'yellowgreen', 'lightcoral']
 

plt.pie(sizes,  labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
